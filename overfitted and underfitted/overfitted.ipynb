#Overfitting :
1.  model performs well on training data but does not perform well on the test data
2. model is memorizing the data instead of learning

```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv ('fruit data.csv')
df
```

```python
X = df.iloc[:, 3:6].values  
Y = df.iloc[:, 0].values
print(X)
print(Y)

```

```python
from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split
X_train, X_test, Y_train , Y_test = train_test_split(X, Y, test_size=.20, random_state= 0)
```

```python
print("splitting done")
print("X train records:", X_train.shape)
print("Y train records:", Y_train.shape)
print("X test records:", X_test.shape)
print("Y test records:", Y_test.shape)
```

```python
from sklearn import tree

clf_2 = tree.DecisionTreeClassifier(min_samples_split=2)

clf_2.fit(X_train, Y_train)

pred_2=clf_2.predict(X_test)
```

```python
print (pred_2)
```

```python
pred_train = clf_2.predict (X_train)
pred_train
```

```python
#from sklearn.metrics import confusion_matrix, precision_score, recall_score , accuracy_score
from sklearn.metrics import accuracy_score
print("Accuracy score on test data: ", accuracy_score(Y_test, pred_2))
print("Accuracy score on training data: ", accuracy_score(Y_train, pred_train))
```

This is problem of Overfitting, We are getting 100% accuracy on training data 
but test data, we are getting 66% accuracy
